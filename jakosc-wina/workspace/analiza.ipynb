```python
wino.corr().sort_values(by = "quality")["quality"].plot(kind = "bar", grid = True, title = "Korelacja z quality")
```

```python
plt.figure(figsize=(16, 9))

for indeks, kolumna in enumerate(wino.columns):
    if kolumna == 'quality':
        continue
    plt.subplot(4, 4, indeks + 1)
    sns.violinplot(x='quality', y=kolumna, data=wino)
```
